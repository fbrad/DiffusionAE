```python
smd_dir = '../processed/SMD/'
```

```python
import numpy as np
import os
```

```python
files = os.listdir(smd_dir)
print(len(files))
for file in files:
    sample = np.load(smd_dir + file)
    max = sample.max()
    min = sample.min()
    print(sample.shape)
    if min == 0 and max == 1:
        print(f'normalized: {file} {min} - {max}')
```
